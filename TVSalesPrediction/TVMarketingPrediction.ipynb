```python
plt.figure(figsize=(12, 6))
# Calculate error terms
errors = y_test - y_pred

# Plotting error terms
plt.scatter(y_test, errors, c='#0B60B0')
plt.axhline(y=0, color='#40A2D8', linestyle='--')  
plt.xlabel('Actual Values', color='#F0EDCF')
plt.ylabel('Error Terms (Actual - Predicted)', color='#F0EDCF')
plt.title('Error Terms Analysis', color='#F0EDCF')
plt.show()

```

```python
plt.figure(figsize=(12, 6))
plt.scatter(y_test, y_pred, s=70, c='#0B60B0', marker='o', label='Actual')
plt.scatter(y_test, y_test, s=50, c='#40A2D8', marker='^', label='Predicted') 
plt.plot(y_test, y_test, color='gray', linewidth=2, label='Perfect Prediction')

plt.xlabel('Actual Values (Y Test)', color='#F0EDCF')
plt.ylabel('Predicted Values', color='#F0EDCF')
plt.title('Actual vs. Predicted Values', color='#F0EDCF')
plt.grid(True)
plt.legend()
plt.show()
```
